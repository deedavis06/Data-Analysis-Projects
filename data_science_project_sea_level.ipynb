```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#Use Pandas to import the data from epa-sea-level.csv.
df=pd.read_csv(r"C:\Users\JoeDe\Downloads\epa-sea-level.csv")
#Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.
x=df['Year']
y=df['CSIRO Adjusted Sea Level']
fig, ax=plt.subplots(figsize=(12,8))
plt.scatter(x,y, s=8)

#Use linregress to plot red line of best fit over scatter plot for the years 1880-2050.
line1=linregress(x,y)
x_adj=np.arange(1880,2050)
y_adj=np.array(df['CSIRO Adjusted Sea Level'])
plt.plot(x_adj, line1.intercept + line1.slope*x_adj, c='r')

#only use data from 2000 to create a new blue line of best fit using linregress
new_df=df.loc[df['Year'] >= 2000]
new_x=new_df['Year']
new_y=new_df['CSIRO Adjusted Sea Level']
line2=linregress(new_x,new_y)
new_x_adj=np.arange(2000,2050)
new_y_adj=np.array(new_df['CSIRO Adjusted Sea Level'])
plt.plot(new_x_adj, line2.intercept + line2.slope*new_x_adj, c='b')

#add title and axis labels
ax.set_title('Rise in Sea Level')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
```

```python
#Find R-squared value of lines of best fit.
print(f"R-squared: {line1.rvalue**2:.6f}")
print(f"R-squared: {line2.rvalue**2:.6f}")
```

```python
print(f"For regression line 1, the intercept of the regression line is: {line1.intercept} and the slope of the regression Line is: {line1.slope} and the Pearson correlation coefficient is: {line1.rvalue} and the pvalue is: {line1.pvalue} and the standard error is: {line1.stderr}")
```
